```python
import pandas as pd

df1 = pd.read_csv("D:\income_averages.tsv", sep="\t")
df1.head()
```

```python
df2 = pd.read_csv("D:\income_gini.tsv", sep="\t")
df2.head()
```

```python
new_dd = pd.merge(df1, df2,  how='inner', on=['country','year'])
```

```python
new_dd
```

```python
# read multiple tsv files and concatenate them in one file

import glob

path = r'D:\income' # use your path
all_files = glob.glob(path + "/*.tsv")

print(all_files)

li = []

for filename in all_files:
    df = pd.read_csv(filename, sep="\t", index_col=None,  header=0)
    li.append(df)

ff = pd.concat(li, axis=0, ignore_index=True)
```

```python
income_countries = ff.drop(['# sptinc992i', '# sptinc992j'], axis=1) 
```

```python
income_countries
```

```python
final = pd.merge(new_dd, income_countries,  how='inner', on=['country','year'])
```

```python
final
```

```python
final.to_csv(r'D:\world_inequality.tsv', sep='\t')
 
```
